```python
# Function to detect outliers using IQR method for each column
def detect_outliers_iqr_df(hrcl_df, columns):
    outliers = {}
    for col in columns:
        q1, q3 = np.percentile(hrcl_df[col], [25, 75])
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr)
        upper_bound = q3 + (1.5 * iqr)
        outliers[col] = hrcl_df[(hrcl_df[col] < lower_bound) | (hrcl_df[col] > upper_bound)][col]
    return outliers
```

```python
# Detect outliers in specified columns
columns_to_check = ['time_spend_company','average_montly_hours','number_project','satisfaction_level','last_evaluation']  # Specify the columns to check for outliers
outliers = detect_outliers_iqr_df(hrcl_df , columns_to_check)
print("Outliers:", outliers)
```

```python
# Remove outliers from specified columns
hrcl_clean = hrcl_df.copy()
for col in columns_to_check:
    hrcl_clean = hrcl_clean[~hrcl_clean[col].isin(outliers[col])]
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        # np.flatnonzero(): Return the indices of the non-zero elements
        
        for candidate in candidates:
            print(f"Model with rank: {i}")
            print(f"Mean validation score: {round(results['mean_test_score'][candidate], 5)} \
                  std: {round(results['std_test_score'][candidate], 5)}")
            print(f"Parameters: {results['params'][candidate]}")
            print("")
```
